```python
############################################
#Order of notebook files usage
#1. splitBlastOut
#2. topBlastResult
#3. removeIdsLessThresh
#4. removeCSVCol
#5. extractFastaSeqfromBlastDB
############################################

#This is for checking and removing which IDs of the topBlast results have less than threshold results.
#For example: if top10Blastout for each ID, how many IDs did not get 25 hits from Blast.


import glob
import os


files=glob.glob("/home/user/thesis/scripts/BlastOut_datasetTMB/BlastOut_datasetTMB_unirefOMBB100/top25BlastSplit/BlastOut_datasetTMB_unirefOMBB100_*.csv")

#empty list
p=[]
#counter to count number of rows/entries that have 25 hits
counter=0

#Here, Thresh or threshold is 27 for example, that will look for 25 hits
#27 because it has two extra rows. Header and query seq ID
Thresh=27
for f in files:
    file = open(f)
    numline = len(file.readlines())
    
    
    if numline==Thresh: 
        
        #counter to count number of rows/entries that have 10 hits
        counter=counter+1
        print(f)
    else:
        
        #append all file names to list before removing them. 
        #Removing that do not have 'Thresh' hits 
        p.append(f)
        
        #Delete the csv file that does not have 25 hits as Blast result 
        os.remove(f)
print(counter)
print(p)
```
